```python
def extract_file(file_path):
        
    print(f"""{file_path} begun.""")

    try:

        df = pd.read_csv(file_path, dtype={
            "tripId": str,
            "UserProgramName": str,
            "UserId": str,
            "UserRole": str,
            "UserCity": str,
            "UserState": str,
            "UserZip": str,
            "UserCountry": str,
            "MembershipType": str,
            "Bike": str,
            "BikeType": str,
            "CheckoutKioskName": str,
            "ReturnKioskName": str,
            "DurationMins": float,
            "AdjustedDurationMins": float,
            "UsageFee": float,
            "AdjustmentFlag": str,
            "Distance": float,
            "EstimatedCarbonOffset": float,
            "EstimatedCaloriesBurned": float,
            "CheckoutDateLocal": str,
            "ReturnDateLocal": str,
            "CheckoutTimeLocal": str,
            "ReturnTimeLocal": str,
            "TripOver30Mins": str,
            "LocalProgramFlag": str,
            "TripRouteCategory": str,
            "TripProgramName": str
            },
            # default encoding induces error
            encoding='latin-1'
            )

        for var in ["CheckoutDateLocal", "ReturnDateLocal"]:
            df[var] = pd.to_datetime(df[var], format="%Y-%m-%d")

        print(f"""{file_path} extracted.""")

    except Exception:
        df=None
        print(f"""{file_path} inread failed.""")

    return df
```

```python
files = os.listdir("./data/BikeShareTrips/")
files = ["./data/BikeShareTrips/" + x for x in files]
```

```python
dfs = [extract_file(x) for x in files]
df = pd.concat(dfs, axis=0)
```
